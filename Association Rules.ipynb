# Importons les librairies et outils necessaires et chargeons nos donn√©es.

```python
import pandas as pd
import numpy as np
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)   
xx=pd.DataFrame(te_ary, columns=te.columns_)
xx.head()
```

# Let's use Apriori Concept

```python
# Support

from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(xx, min_support=0.6, use_colnames=True) 
frequent_itemsets
```

```python
# Confidence

from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
# Lift

association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```

```python
column = []  
for i in range(0, 7501):  
    column.append([str(data.values[i,j]) for j in range(0, 20)])
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_ary = te.fit(column).transform(column)
data = pd.DataFrame(te_ary, columns=te.columns_)
data
```

```python
apriori(data, min_support=0.0045)
```

```python
frequent_itemsets = apriori(data, min_support=0.0045, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.007)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="lift", min_threshold=1.25)
```
